4. What is a box plot, and what can it tell you about the distribution of data?
 - A box plot, also known as a box-and-whisker plot, is a graphical representation of a dataset that displays its central tendency, variability, and distribution. It is based on a five-number summary of the data, which includes:  
 Minimum – the smallest data value (excluding outliers)
First Quartile (Q1) – the 25th percentile
Median (Q2) – the 50th percentile or middle value
Third Quartile (Q3) – the 75th percentile
Maximum – the largest data value (excluding outliers)
In a box plot:
A rectangle (box) is drawn from Q1 to Q3, representing the interquartile range (IQR), which contains the middle 50% of the data.
A line inside the box shows the median (Q2).
Whiskers extend from the box to the minimum and maximum values within 1.5 times the IQR from the quartiles.
Outliers, if any, are shown as individual points beyond the whiskers.
Interpretation:
A box plot provides insights into the distribution of the data:
It shows the spread (variability) through the length of the box and whiskers.
It indicates central tendency through the position of the median.
It helps detect skewness:
If the median is closer to Q1, the data is right-skewed.
If it is closer to Q3, the data is left-skewed.
It highlights the presence of outliers, which are unusually high or low values.

5. Discuss the role of random sampling in making inferences about populations?
 - Random sampling is a fundamental technique in statistics used to make inferences about a population based on a subset (sample) of that population. It plays a crucial role in ensuring that the conclusions drawn are valid, reliable, and generalizable.
 * Definition:  
Random sampling is a process where every individual in the population has an equal chance of being selected in the sample. This eliminates selection bias and allows for the creation of a representative sample.
 * Why Is Random Sampling Important?  
 . Ensures Representativeness:  
Random samples reflect the characteristics of the overall population, allowing researchers to make accurate generalizations.  
. Reduces Bias:  
Since the selection is not based on personal judgment or patterns, it helps avoid sampling bias and promotes fairness.  
. Supports Valid Inferences:  
Random sampling provides a basis for applying probability theory, which enables statisticians to calculate margins of error and confidence intervals.  
.Allows Hypothesis Testing:  
It is essential for performing statistical tests (like t-tests or chi-square tests), which are used to make decisions about population parameters.


6.  Explain the concept of skewness and its types. How does skewness affect the interpretation of data?
 - Skewness: Concept and Definition  
Skewness is a statistical measure that describes the asymmetry of a data distribution around its mean. In a perfectly symmetrical distribution, the left and right sides of the graph are mirror images, and the mean, median, and mode are equal.
When the distribution is not symmetrical, it is said to be skewed.
* ypes of Skewness  
a. Positive Skewness (Right Skewed):  
.The right tail (higher values) is longer.  
.Most values are concentrated on the lower end of the scale.  
.The mean is greater than the median, and the median is greater than the mode.  
*Example: Income distribution, where a few people earn significantly more than the rest.  
b. Negative Skewness (Left Skewed):   
.The left tail (lower values) is longer.    
.Most values are concentrated on the higher end of the scale.    
.The mean is less than the median, and the median is less than the mode.  
*Example: Age at retirement, where most people retire at a later age but a few retire early.  
c.Zero Skewness (Symmetrical Distribution):    
.The distribution is evenly balanced on both sides of the mean.    
.The mean, median, and mode are approximately equal.  
*Example: Ideal normal distribution (bell curve).      
*Effect of Skewness on Data Interpretation
 .Central Tendency:  
In skewed data, the mean is affected by extreme values (outliers), so it may not represent the "typical" value.
The median is often a better measure of central tendency for skewed data.
 .Statistical Analysis:  
Many statistical tests assume normal distribution (zero skewness).  
If the data is skewed, it may be necessary to use data transformation or non-parametric tests.  
 .Visual Interpretation:    
Skewness affects how a distribution looks in graphs like histograms and box plots.    
In a box plot, skewness can be detected if the median line is not centered, or if one whisker is longer.    
.Real-Life Decision Making:    
Understanding skewness helps in making better decisions. For example, in positively skewed data like salaries, using the median gives a more accurate picture of what most people earn.

8.  Discuss the conditions under which the binomial distribution is used?
 - Conditions for Using the Binomial Distribution
A binomial distribution is appropriate when the following four conditions are met:  
a. Fixed Number of Trials (n):  
There must be a set number of trials or experiments.  
 .Example: Tossing a coin 10 times (n = 10).  
b. Only Two Possible Outcomes:  
Each trial must result in exactly one of two outcomes:  
Commonly labeled as "success" and "failure".
.Example: A student passes or fails an exam.  
c. Constant Probability of Success (p):  
The probability of success remains the same for each trial.  
.Example: If the probability of getting heads on a fair coin is 0.5, it remains 0.5 for every toss.  
d. Independent Trials:  
Each trial must be independent of the others.  
The outcome of one trial does not affect the outcome of another.

9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule)?
 - Properties of the Normal Distribution  
. Symmetry  
The distribution is perfectly symmetrical about the mean.
This means the left and right sides of the curve are mirror images.    
. Mean = Median = Mode
All three measures of central tendency are equal and located at the center of the distribution.  
. Bell-Shaped Curve  
The curve rises to a peak at the mean and falls off equally on both sides.  
. Asymptotic to the X-Axis  
The tails of the curve approach, but never touch, the horizontal axis.
This implies that extreme values are possible but rare.  
. Defined by Two Parameters  
Mean (μ): Determines the center (location) of the distribution.
Standard Deviation (σ): Measures the spread or dispersion of the distribution.
. Total Area Under the Curve = 1  
This represents the total probability (100%).  
* Empirical Rule (68-95-99.7 Rule)
The empirical rule provides a quick estimate of the spread of data in a normal distribution. It states that:  
.68% of the data falls within 1 standard deviation of the mean
(μ − σ to μ + σ)  
.95% of the data falls within 2 standard deviations of the mean
(μ − 2σ to μ + 2σ)    
.99.7% of the data falls within 3 standard deviations of the mean
(μ − 3σ to μ + 3σ)

11.  Explain what a random variable is and differentiate between discrete and continuous random variables ?
 - What is a Random Variable?
A random variable is a numerical value that represents the outcome of a random experiment.

It is not random itself, but the value it takes depends on chance.

It is usually denoted by capital letters like X, Y, Z.

A random variable maps outcomes of a random process to numbers so they can be analyzed statistically.

Types of Random Variables
a. Discrete Random Variable:
A discrete random variable can take on a countable number of distinct values.

These values are often whole numbers.

Arise from counting processes.

Examples:
Number of heads in 3 coin tosses → X ∈ {0, 1, 2, 3}

Number of students in a class

Number of emails received in an hour

Key Characteristics:
Values can be listed (finite or countably infinite)

Has a probability mass function (PMF)

b. Continuous Random Variable:
A continuous random variable can take on any value within a given range, including fractions and decimals.

Arise from measuring processes.

Examples:
Height of a person

Time taken to finish a task

Temperature in a city

Key Characteristics:
Infinite number of possible values in an interval

Has a probability density function (PDF)

The probability of the variable taking an exact value is zero; instead, we consider intervals
